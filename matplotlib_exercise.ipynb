```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvは.hist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
# CSVファイルを読み込む
file_path = 'data/hist_data.csv'  
data = pd.read_csv(file_path)

# ヒストグラムをプロット
plt.figure(figsize=(8, 6))
plt.hist(data['aデータ'], bins=30, alpha=0.5, label='aデータ')
plt.hist(data['bデータ'], bins=30, alpha=0.5, label='bデータ')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
plt.savefig('./imgs/matplotlib/plot_1.png')  # プロットを保存
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# 日本語フォントを有効にする
japanize_matplotlib.japanize()

# CSVファイルを読み込む
file_path = 'data/scatter_data.csv'
data = pd.read_csv(file_path)

# データフレームのカラム名を表示して確認
print(data.columns)

# 散布図をプロット
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], label='データポイント')  # 最初の2カラムをプロット
plt.xlabel('x軸')
plt.ylabel('y軸')
plt.title('xとyの散布図')
plt.legend()
plt.savefig('./imgs/matplotlib/plot_2.png')  # プロットを保存
plt.show()
```
