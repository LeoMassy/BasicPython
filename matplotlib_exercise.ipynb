## 1問目
使うcsvは.hist_data.csvです。下にコードを書いてください。

```python
# CSVファイルを読み込む
file_path = 'data/hist_data.csv'  
data = pd.read_csv(file_path)

# ヒストグラムをプロット
plt.figure(figsize=(8, 6))
plt.hist(data['aデータ'], bins=30, alpha=0.5, label='aデータ')
plt.hist(data['bデータ'], bins=30, alpha=0.5, label='bデータ')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend()
plt.savefig('./imgs/matplotlib/plot_1.png')  # プロットを保存
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルを読み込む
file_path = 'data/scatter_data.csv'
data = pd.read_csv(file_path)

# データフレームのカラム名を表示して確認
print(data.columns)

# データを3つのグループに分ける
group1 = data.iloc[:len(data)//3]
group2 = data.iloc[len(data)//3:2*len(data)//3]
group3 = data.iloc[2*len(data)//3:]

# 散布図をプロット
plt.figure(figsize=(8, 6))
plt.scatter(group1.iloc[:, 0], group1.iloc[:, 1], color='blue', label='グループ1')
plt.scatter(group2.iloc[:, 0], group2.iloc[:, 1], color='red', label='グループ2')
plt.scatter(group3.iloc[:, 0], group3.iloc[:, 1], color='green', label='グループ3')
plt.xlabel('x軸')
plt.ylabel('y軸')
plt.title('散布図')
plt.legend()
plt.savefig('./imgs/matplotlib/plot_2.png')  # プロットを保存
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

```python
# CSVファイルを読み込む
file_path = 'data/line_data.csv'
data = pd.read_csv(file_path)

# データフレームのカラム名を表示して確認
print(data.columns)

# 折れ線グラフをプロット
plt.figure(figsize=(10, 6))
plt.plot(data['期間'], data['件数'], marker='o', linestyle='-', color='blue', label='件数')

# グラフのタイトルと軸ラベルを設定
plt.title('期間と件数の折れ線グラフ')
plt.xlabel('期間')
plt.ylabel('件数')

# 凡例とグリッドを追加
plt.legend()
plt.grid(True)

# グラフを画像として保存
plt.savefig('./imgs/matplotlib/plot_3_1.png')

# グラフを表示
plt.show()

```

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
# CSVファイルを読み込む
file_path = 'data/line_data.csv'
data = pd.read_csv(file_path)

# データフレームのカラム名を表示して確認
print("カラム名:", data.columns)
# '期間' カラムのユニークな値を表示して確認
print("期間のユニークな値:", data['期間'].unique())

# データフレームの確認
print(data.head())

# '期間' カラムを文字列型として扱う
data['期間'] = data['期間'].astype(str)

# プロットを作成
plt.figure(figsize=(12, 8))

# 棒グラフの位置調整のためのオフセット
offset = 0.2
x_positions = range(len(data['期間']))  # インデックスの位置を使う
x_positions_件数 = [x - offset for x in x_positions]
x_positions_費用 = [x + offset for x in x_positions]



# '費用'の棒グラフ
plt.bar(x_positions_費用, data['費用'], width=0.4, color='red', label='費用')

# x 軸のラベルを '期間' の値に設定
plt.xticks(x_positions, data['期間'])

# グラフのタイトルと軸ラベルを設定
plt.title('期間ごとの費用の棒グラフ')
plt.xlabel('期間')
plt.ylabel('値')

# 凡例とグリッドを追加
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# グラフを画像として保存
plt.savefig('./imgs/matplotlib/plot_3_2.png')

# グラフを表示
plt.show()
```
